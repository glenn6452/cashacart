```python
query = open('cashacart.sql','r')
```

```python
d1 = pd.read_sql_query(query.read(), sql_engine)
#d1.head()
```

```python
d2 = pd.read_csv('split.csv')
#d2.head()
```

```python
df = pd.merge(d1,d2, left_on='pos_code', right_on='pos_code')
#df.head()
```

```python
sasm = df['sasm'].sort_values().unique()
#sasm
```
