````python
connect_string = 'mysql://cashacart_monitor:aabb1122@prod-rep3.cfti8umsjnvm.ap-southeast-1.rds.amazonaws.com/trade'
sql_engine = sql.create_engine(connect_string)
````

````python
query = open('cashacart.sql','r')
````

````python
d1 = pd.read_sql_query(query.read(), sql_engine)
#d1.head()
````

````python
d2 = pd.read_csv('split.csv')
#d2.head()
````

````python
df = pd.merge(d1,d2, left_on='pos_code', right_on='pos_code')
#df.head()
````

````python
reg_sign = pd.DataFrame(df.groupby(['region'])['sign'].sum().reset_index()).sort_values('sign')
reg_apps = pd.DataFrame(df.groupby(['region'])['apps'].sum().reset_index()).sort_values('apps')
sasm_sign = pd.DataFrame(df.groupby(['sasm'])['sign'].sum().reset_index()).sort_values('sign')
sasm_apps = pd.DataFrame(df.groupby(['sasm'])['apps'].sum().reset_index()).sort_values('apps')
asm_sign = pd.DataFrame(df.groupby(['asm'])['sign'].sum().reset_index()).sort_values('sign')
asm_apps = pd.DataFrame(df.groupby(['asm'])['apps'].sum().reset_index()).sort_values('apps')
````

**Per Region**

````python
ax = sns.barplot(x="apps", y='region', data=reg_apps,palette="Set2")
ax.set_alpha(0.8)
ax.set_title("Applications per Region", fontsize=14)
ax.set_xlabel("Apps Apps Apps", fontsize=13);
ax.set_ylabel("Region", fontsize=13);
#ax.set_xticks([0,2.5,5,7.5,10,12.5])

# create a list to collect the plt.patches data
totals = []

# find the values and append to list
for i in ax.patches:
    totals.append(i.get_width())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+.1, i.get_y()+.28, \
            str(round(i.get_width(), 2)), fontsize=13,
color='#515A5A')
# invert for largest on top 
ax.invert_yaxis()
#plt.savefig('test.png', bbox_inches = 'tight',dpi = 300)
````
