```python
import pandas as pd
import sqlalchemy as sql
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
connect_string = 'mysql://cashacart_monitor:aabb1122@prod-rep3.cfti8umsjnvm.ap-southeast-1.rds.amazonaws.com/trade'
sql_engine = sql.create_engine(connect_string)
```

```python
query = open('query.sql','r')
```

```python
df = pd.read_sql_query(query.read(), sql_engine)
```

```python
#df.dtypes
```

```python
apps_daily = pd.DataFrame(df.groupby(['retailer','dateApplied','pos_name'])['apps'].sum().groupby(level=0).mean())
apps_daily = apps_daily.reset_index().sort_values('apps', ascending=False)
signed_daily = pd.DataFrame(df.groupby(['retailer','dateApplied','pos_name'])['sign'].sum().groupby(level=0).mean())
signed_daily = signed_daily.reset_index().sort_values('sign', ascending=False)
```

```python
ax = sns.barplot(x="apps", y="retailer", data=apps_daily.head(10).sort_values('apps'),palette="bone_r")
ax.set_alpha(0.8)
ax.set_title("Average Daily Apps per POS of Partners", fontsize=14)
ax.set_xlabel("Average Daily Apps of POS", fontsize=13);
ax.set_ylabel("Partners", fontsize=13);
ax.set_xticks([0,2.5,5,7.5,10,12.5])

# create a list to collect the plt.patches data
totals = []

# find the values and append to list
for i in ax.patches:
    totals.append(i.get_width())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+.1, i.get_y()+.28, \
            str(round(i.get_width(), 2)), fontsize=11,
color='#515A5A')
# invert for largest on top 
ax.invert_yaxis()
#plt.savefig('test.png', bbox_inches = 'tight',dpi = 300)
```

```python
retailer_day_pos = df.groupby(['retailer','dateApplied','pos_name']).sum().reset_index()
```

```python
retailer_day_pos['ar'] = retailer_day_pos['appr']/(retailer_day_pos['apps']-retailer_day_pos['canc_pre_result'])
```

```python
ar_daily = pd.DataFrame(retailer_day_pos.groupby(['retailer','dateApplied','pos_name'])['ar'].sum().groupby(level=0).mean())
ar_daily = ar_daily.reset_index().sort_values('ar', ascending=False)
```

```python
ax = sns.barplot(x="ar", y="retailer", data=ar_daily.head(10).sort_values('ar'),palette='PuRd')
ax.set_alpha(0.8)
ax.set_title("Top 10 Average AR per POS of Partners", fontsize=14)
ax.set_xlabel("Average AR of POS", fontsize=13);
ax.set_ylabel("Partners", fontsize=13);
ax.set_xticks([0,.5,1])

# create a list to collect the plt.patches data
totals = []

# find the values and append to list
for i in ax.patches:
    totals.append(i.get_width())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+.03, i.get_y()+.33, \
            str(round(i.get_width()*100, 2)) + '%', fontsize=11,
color='#515A5A')
# invert for largest on top 
ax.invert_yaxis()
#plt.savefig('test3.png', bbox_inches = 'tight',dpi = 300)
```
